```python
import nltk
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
example_string = """ Muad'Dib learned rapidly because his first training was in how to learn.
And the first lesson of all was the basic trust that he could learn. It's shocking to find how
many people do not believe they can learn, and how many more believe learning to be
difficult."""
print(sent_tokenize(example_string))
print(word_tokenize(example_string))
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
worf_quote = "If you’re familiar with the basics of using Python and would like to get your feet wet with NLP "
words_in_quote = word_tokenize(worf_quote)
print(words_in_quote)
stop_words = set(stopwords.words("english"))
filtered_list = []
for word in words_in_quote:
    if word.casefold() not in stop_words:
        filtered_list.append(word)
        filtered_list = [word for word in words_in_quote if word.casefold() not in stop_words]
print(filtered_list)
```
