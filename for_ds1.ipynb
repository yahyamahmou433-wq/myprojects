```python
#EX1 
#strings:
#1.Write a function to count the number of vowels in a given string
def number_of_vowels(text):
  Vowels=["a","e","i","u","o"]
  cont=0
  text=text.lower()
  for i in text:
    if i in Vowels:
      cont+=1
  print(cont)
number_of_vowels("yahya loves python and python is easy and python is popular language and python is good for data science")
```
